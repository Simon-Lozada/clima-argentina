```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("cordoba_masticado.csv")
print(df.columns)
```

```python
df["Temperatura media anual"] = pd.to_numeric(df["Temperatura media anual"]) 
df["Temperatura máxima media anual"] = pd.to_numeric(df["Temperatura máxima media anual"]) 
df["Temperatura mínima media anual"] = pd.to_numeric(df["Temperatura mínima media anual"]) 
df["Precipitación total anual de lluvia y/o nieve derretida (mm)"] = pd.to_numeric(df["Precipitación total anual de lluvia y/o nieve derretida (mm)"]) 
df["Velocidad media anual del viento (Km/h)"] = pd.to_numeric(df["Velocidad media anual del viento (Km/h)"])
df["Total días con lluvia durante el año"] = pd.to_numeric(df["Total días con lluvia durante el año"]) 
df["Total días que nevó durante el año"] = pd.to_numeric(df["Total días que nevó durante el año"]) 
df["Total días con tormenta durante el año"] = pd.to_numeric(df["Total días con tormenta durante el año"]) 
df["Total días con niebla durante el año"] = pd.to_numeric(df["Total días con niebla durante el año"]) 
df["Total días con tornados o nubes de embudo durante el año"] = pd.to_numeric(df["Total días con tornados o nubes de embudo durante el año"]) 
df["Total días con granizo durante el año"] = pd.to_numeric(df["Total días con granizo durante el año"])
df.info()
df.head(10)
```

```python
#Aca declaro las variables en base a las filas del dataset
año = df["Año"]
tempmedia = df["Temperatura media anual"]
tempmax = df["Temperatura máxima media anual"]
tempmin = df["Temperatura mínima media anual"]
lluvias__nieve = df["Precipitación total anual de lluvia y/o nieve derretida (mm)"]
velviento = df["Velocidad media anual del viento (Km/h)"]
lluvia = df["Total días con lluvia durante el año"]
nevadas = df["Total días que nevó durante el año"]
tormentas = df["Total días con tormenta durante el año"]
niebla = df["Total días con niebla durante el año"]
tornados_nubes = df["Total días con tornados o nubes de embudo durante el año"]
granizo = df["Total días con granizo durante el año"]
```

```python
#aqui hago un grafico de regresió lineal de las temperaturas maximas, medias y mínimas
sns.regplot(x = año, y = tempmedia, data = df ).set(title = "Comparacion entre la temperatura minima, maxima y media ")
sns.regplot(x = año, y = tempmax, data = df)
sns.regplot(x = año, y = tempmin, data = df)
```
